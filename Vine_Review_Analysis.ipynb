```python
import pandas as pd
```

```python
vine_table_df=pd.read_csv("Resources/vine_table.csv")
vine_table_df.head()
```

```python
print(f"Size:\n")
print(f'{vine_table_df.shape}\n')
print("Data types: \n")
print(vine_table_df.dtypes)
```

```python
#1. Filter df to where total_votes>20
vt_votes_over20=vine_table_df.query("total_votes>=20")
vt_votes_over20.head()
```

```python
#2. Filter df to retrieve all the rows where helpful_votes/total_votes >= 50%
vt_helful_over50=vt_votes_over20.query("helpful_votes/total_votes >= .5")
vt_helful_over50.head()
```

```python
#3. Filter df to where vine==Y
vt_vine_y=vt_helful_over50.query("vine =='Y'")
vt_vine_y.head()
```

```python
#4. Filter df to where vine==N
vt_vine_n=vt_helful_over50.query("vine =='N'")
vt_vine_n.head()
```

```python
#5. Determine the total number of reviews, 
#the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
#PAID
number_of_reviews_paid=vt_vine_y["review_id"].count()
number_of_5star_reviews_paid=vt_vine_y["star_rating"][(vt_vine_y.star_rating ==5)].count()
percentage_5star_paid=vt_vine_y["star_rating"][(vt_vine_y.star_rating ==5)].count()/vt_vine_y["star_rating"].count()

#UNPAID
number_of_reviews_unpaid=vt_vine_n["review_id"].count()
number_of_5star_reviews_unpaid=vt_vine_n["star_rating"][(vt_vine_n.star_rating ==5)].count()
percentage_5star_unpaid=vt_vine_n["star_rating"][(vt_vine_n.star_rating ==5)].count()/vt_vine_n["star_rating"].count()


output=f'-------PAID------------------------------\n \
      Total number of reviews: {number_of_reviews_paid} \n \
      Number of 5-star reviews: {number_of_5star_reviews_paid}\n \
      Percentage of 5-reviews: {percentage_5star_paid:.2%}\n \
\n-------UNPAID------------------------------\n \
      Total number of reviews: {number_of_reviews_unpaid} \n \
      Number of 5-star reviews: {number_of_5star_reviews_unpaid}\n \
      Percentage of 5-reviews: {percentage_5star_unpaid:.2%}\n \
      '
print(output)
```

```python
with open('results.txt', 'w') as f:
    f.write(output)
```
